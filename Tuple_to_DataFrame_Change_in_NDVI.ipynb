```python
list_to_append_new_values = []
heading = ('Geometry','Mean' ,'Positive Delta NDVI', 'Negative Delta NDVI', 'Positive Percent Delta NDVI', 'Negative Percent Delta NDVI')
list_to_append_new_values.append(heading)
for i in range(len(unique_points)):
    new_df = pd.DataFrame(df[df['.geo'] == unique_points[i]])
    new_df['Delta NDVI'] = new_df['NDVI'].diff(1)
    new_df.fillna(0)
    mean = new_df['NDVI'].mean(axis = 0)
    ct_total = len(new_df['Delta NDVI'])
    ct_positive = len(new_df.loc[new_df['Delta NDVI'] > 0])
    ct_negative = len(new_df.loc[new_df['Delta NDVI'] < 0])
    positive_percent=round((ct_positive/ct_total)*100,3)
    negative_percent=round((ct_negative/ct_total)*100,3)
    list_to_append_new_values.append((unique_points[i],mean, ct_positive, ct_negative, positive_percent, negative_percent))
list_to_append_new_values
```

```python
tuple_to_dataframe = pd.DataFrame(list_to_append_new_values, columns=['Geometry','Mean' ,'Positive_Delta_NDVI', 'Negative_Delta_NDVI', 'Positive_Percent_Delta_NDVI (%)', 'Negative_Percent_Delta_NDVI (%)'])
tuple_to_dataframe = tuple_to_dataframe[1:]
```

```python
tuple_to_dataframe['Geometry'] = tuple_to_dataframe['Geometry'].str.strip('{"type":"Point","coordinates":}')
tuple_to_dataframe
```

```python
tuple_to_dataframe.dtypes
```

### Categorical to numerical

```python
tuple_to_dataframe['Mean'] = tuple_to_dataframe['Mean'].astype(str).astype(float)
tuple_to_dataframe['Positive_Delta_NDVI'] = tuple_to_dataframe['Positive_Delta_NDVI'].astype(str).astype(float)
tuple_to_dataframe['Negative_Delta_NDVI'] = tuple_to_dataframe['Negative_Delta_NDVI'].astype(str).astype(float)
tuple_to_dataframe['Positive_Percent_Delta_NDVI (%)'] = tuple_to_dataframe['Positive_Percent_Delta_NDVI (%)'].astype(str).astype(float)
tuple_to_dataframe['Negative_Percent_Delta_NDVI (%)'] = tuple_to_dataframe['Negative_Percent_Delta_NDVI (%)'].astype(str).astype(float)
print(tuple_to_dataframe.dtypes)
```
