### 2. Delta NDVI which subtracts NDVI values between rows


```python
print("The differnce between NDVI for the geometry",str1,"on two separate acquistions/dates are: ")
p1['Delta NDVI'] = p1['NDVI'].diff(1)
p1.fillna(0)
```

### 3. Find the count of positive and negative values in "Delta NDVI" column per acquisition and project it as "Cumulative Positive Change" and "Cumulative Negative Change"
### 4. Calculate the percentage of positive and negative values in "Delta NDVI" column per acquisition and project it as "Cumulative Positive Change Percent" and "Cumulative Negative Change Percent".

```python
ct_total = len(p1['Delta NDVI'])
ct_positive = len(p1.loc[p1['Delta NDVI'] > 0])
ct_negative = len(p1.loc[p1['Delta NDVI'] < 0])
positive_percent=round((ct_positive/ct_total)*100,6)
negative_percent=round((ct_negative/ct_total)*100,6)
print("\n"," - "*20)
print('\nTotal number of rows in the column Delta NDVI:', ct_total)
print("\n"," - "*20)
print("\nCumulative 'Positive' Change in Delta NDVI is:", ct_positive)
print("\n"," - "*20)
print("\nCumulative 'Negative' Change in Delta NDVI is:", ct_negative)
print("\n"," - "*20)
print('\nCumulative Positive Change Percent in Delta NDVI is:', positive_percent, "%")
print("\n"," - "*20)
print('\nCumulative Negative Change Percent in Delta NDVI is:', negative_percent, "%")
print("\n"," - "*20)
```

```python
a = 100-positive_percent
values = [positive_percent, a]
fig1, ax1 = plt.subplots(figsize = (5,5))
theme = plt.get_cmap('spring_r')
ax1.set_prop_cycle("color", [theme(1. * i / len(values))
                             for i in range(len(values))])
plt.pie(values, startangle = 90, autopct = "%0.3f%%")
plt.title('The Cumulative Positive Change Percent in Delta NDVI in yellow')
plt.show()
```

```python
b = 100-negative_percent
values = [negative_percent, b]
fig1, ax1 = plt.subplots(figsize = (5,5))
theme = plt.get_cmap('nipy_spectral_r')
ax1.set_prop_cycle("color", [theme(1. * i / len(values))
                             for i in range(len(values))])
plt.pie(values, startangle = 90, autopct = "%0.3f%%")
plt.title('The Cumulative Negative Change Percent in Delta NDVI in Grey')
plt.show()
```
